```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('team-info.csv', usecols=['team_name', 'ast_pg'])
rows_to_include = [6, 21, 19, 7, 13, 5, 9, 3]
data = data.loc[rows_to_include]

fig = plt.figure(figsize=(9,6))
plt.plot(data['team_name'], data['ast_pg'], color='red')
plt.title('Assists per game of the Top 8 Teams in the Western Conference')
plt.ylabel('Assists per game')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

data=pd.read_csv('team-info.csv', usecols=['team_name', 'pts_pg'])
rows_to_include = [12, 1, 18, 2, 17, 4, 16, 25]
data = data.loc[rows_to_include]

x = data['team_name']
y = data['pts_pg']
colors = ['darkgreen', 'green', 'blue', 'maroon', 'yellow', 'red', 'orange', 'gray']
plt.bar(x,y, color=colors)
plt.ylim(100, 120)
plt.ylabel('Points per game')
plt.title('Points per game of the Top 4 and the Bottom 4 Teams of the Eastern Conference')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('team-info.csv', usecols=['team_name', 'pts_pg', 'opp_pts_pg'], nrows= 15)

fig = plt.figure(figsize=(15, 10))
plt.plot(data['team_name'], data['pts_pg'], color='blue', label='Points per game')
plt.plot(data['team_name'], data['opp_pts_pg'], color = 'red', label="Opponent's Points per game")
plt.ylabel('Points per game', fontsize=14)
plt.legend()
plt.show()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6
import pandas as pd

data=pd.read_csv('team-info.csv')

output_file("team-info.html")
source = ColumnDataSource(data)

hover = HoverTool(tooltips=[
                  ("Assists per game", "@ast_pg"),
                  ("Total Wins", "@w")
                  ])

p = figure(title="Correlation of Assits per game with Total Wins")
p.add_tools(hover)
p.xaxis.axis_label = "Assists per game"
p.yaxis.axis_label = 'Total Wins'

p.circle('ast_pg', 'w', size=10, source=source)

show(p)
```

For the second graph, I used a bar graph to display the points per game of the top 4 teams compared to the bottom 4 teams in the eastern conference. The graphic shows a trend that high ranking teams record a higher points per game.
For the next graph, I used another line graph comparing the team's total points compared to their opponents. It shows that teams at the bottom of the ranking such as the Rockets have a huge gap in points to their opponents. For this graph I increased the canvas size of the figure to fit a total of 15 teams into the graph.
For the last graph, I used a scatter plot to determien if there is any connection between the total wins recorded and the team's respective assists per game. As seen in the graph there are no trends or signs that it has any correlation as the points are scattered randomly.
